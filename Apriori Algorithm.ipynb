```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data preprocessing
df = pd.read_csv("Market_Basket_Optimisation.csv", header=None)
df.head()
```

```python
# Replacing NaNwith zeros
df.fillna(0,inplace=True)
df.head()
```

```python
#Further Data Pre-processing
transactions = []
for i in range (0,len(df)): 
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str (df.values[i,j])!="0"])
# convert to string for all column ranging from 0-19, yang value not 0
```

```python
transactions[0]
```

```python
#Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions,min_support = 0.003, min_confidence = 0.2, min_lift = 3, max_length = 3)
results = list(rules)
# apriori(input=transactions:must be in list format)
```

```python
#save trasnsctions and results as csv
df1 = pd.DataFrame(transactions)
df1.to_csv("transactions.csv")
df2=pd.DataFrame(results)
df2.to_csv("results.csv")

results=pd.read_csv('results.csv')
results.head(10)
```
