```python
# temperature exercise 
# histogram - temperature

#import libraries
import matplotlib.pyplot as plt
import pandas as pd

#import dataset
ds = pd.read_csv('temperature.csv', header=None)
# x=ds.iloc[:,:].values
#x=ds.iloc[:,[0]].values
x=ds.iloc[:,0:1].values
```

```python
ds.info()
```

```python
# plt.hist(x, bins=20, color='black') 
plt.hist(x, color='pink') 
plt.title("Distribution of average monthly temperature")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show() 
```
