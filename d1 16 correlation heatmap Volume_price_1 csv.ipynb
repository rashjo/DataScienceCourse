```python
# pg 147
import pandas as pd
df = pd.read_csv('Volume_price_1.csv')
corr = df.corr()
print(corr)
```

```python
# read file into a pandas dataframe
from pandas import DataFrame, read_csv
df=read_csv('Volume_price_1.csv')
print(df)
```

```python
# pg 148
# use color scale to represent
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

```python
# use color scale to represent: cmap=
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr, cmap="ocean_r",
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```


tengok scale di sebelah kiri.
- the darker is close to 1
- the lighter is close to min value
